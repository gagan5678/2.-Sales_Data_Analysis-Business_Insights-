# Data Analysis

### Data Quality Check

```python
print('\n Missing Values:\n')
missing_data = df.isnull().sum()
print(missing_data[missing_data>0])

print(f'\n Duplicate Rows:{df.duplicated().sum()}')
```

# Data Type conversion 

```python
df['Order Date']= pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Ship Date']= pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')

print(f'Order Date: {df['Order Date'].dtype}')
print(f'Ship Date: {df['Ship Date'].dtype}')
```

```python
df.info()
```

## Univariate Analysis

```python
categorical_cols = ['Customer Segment','Category','Region','Payment Mode','Delivery Status','Auto Reorder']
for col in categorical_cols:
    print(f'\n--{col}--')
    print(df[col].value_counts())

    plt.figure(figsize=(6,6))
    df[col].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation =30)
    plt.tight_layout()
    plt.show()
```

## Profit Margin Analysis

```python
df['Profit Margin']= (df['Profit']/df['Sales Amount'])*100

# By Category
pm_category = df.groupby('Category')['Profit Margin'].mean().round(2)
print("\nProfit Margin by Category:\n", pm_category)

# By Region
pm_region = df.groupby('Region')['Profit Margin'].mean().round(2)
print("\nProfit Margin by Region:\n", pm_region)

pm_category.plot(kind='bar', title='Profit Margin by Category')
plt.xticks(rotation=30)
plt.ylabel('Profit Margin %')
plt.show()
```

## Numerical Variable Analysis

```python
numerical_cols = ['Quantity','Unit Price','Discount (%)','Sales Amount',
                  'Cost Price','Profit','Stock Left','Reorder Quantity']
for col in numerical_cols:
    print(f'\n---{col}---')
    print(f'Mean: {df[col].mean():.2f}')
    print(f'Median: {df[col].median():.2f}')
    print(f'Std Dev: {df[col].std():.2f}')
    print(f'Min: {df[col].min():.2f}')
    print(f'Max: {df[col].max():.2f}')

    plt.figure(figsize=(8,4))

    plt.subplot(1,2,1)
    df[col].hist(bins=30, color='lightgreen',alpha=0.7)
    plt.title(f"Distribution of {col}")

    plt.subplot(1,2,2)
    df.boxplot(column=col)
    plt.title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()
```

## Bivariate Analysis

```python
category_performance = df.groupby('Category').agg({'Sales Amount':'sum',
                                                   'Profit': 'sum', 
                                                   'Order ID':'count'}).rename(columns={'Order ID':'Order Count'})

category_performance['Profit Margin']=(category_performance['Profit']/category_performance['Sales Amount'])*100
print(category_performance,'\n\n')

import matplotlib.pyplot as plt

# Sales Amount
plt.figure(figsize=(6,4))
category_performance['Sales Amount'].plot(kind='bar')
plt.title('Sales Amount by Category')
plt.xticks(rotation=30)
plt.ylabel('Sales')
plt.show()

# Profit
plt.figure(figsize=(6,4))
category_performance['Profit'].plot(kind='bar', color='orange')
plt.title('Profit by Category')
plt.xticks(rotation=30)
plt.ylabel('Profit')
plt.show()

# Order Count
plt.figure(figsize=(6,4))
category_performance['Order Count'].plot(kind='bar', color='green')
plt.title('Order Count by Category')
plt.xticks(rotation=30)
plt.ylabel('Orders')
plt.show()

# Profit Margin
plt.figure(figsize=(6,4))
category_performance['Profit Margin'].plot(kind='bar', color='red')
plt.title('Profit Margin (%) by Category')
plt.xticks(rotation=30)
plt.ylabel('Profit %')
plt.show()
```

```python
region_performance = df.groupby('Region').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Order ID': 'count'
}).rename(columns={'Order ID': 'Order Count'})

region_performance['Profit Margin'] = (
    region_performance['Profit'] / region_performance['Sales Amount']
) * 100

print(region_performance, '\n\n')

plt.figure(figsize=(12,8))

# Sales Amount
plt.subplot(2,2,1)
region_performance['Sales Amount'].plot(
    kind='pie', autopct='%1.1f%%', startangle=90
)
plt.title('Sales Distribution by Region')
plt.ylabel('')

# Profit
plt.subplot(2,2,2)
region_performance['Profit'].plot(
    kind='pie', autopct='%1.1f%%', startangle=90
)
plt.title('Profit Distribution by Region')
plt.ylabel('')

# Order Count
plt.subplot(2,2,3)
region_performance['Order Count'].plot(
    kind='pie', autopct='%1.1f%%', startangle=90
)
plt.title('Order Distribution by Region')
plt.ylabel('')

# Profit Margin
plt.subplot(2,2,4)
region_performance['Profit Margin'].plot(
    kind='pie', autopct='%1.1f%%', startangle=90
)
plt.title('Profit Margin Distribution by Region')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
df['Order Month'] = df['Order Date'].dt.to_period('M')

# Monthly Trends
monthly_trends = df.groupby('Order Month').agg({
    'Sales Amount': 'sum',
    'Profit': 'sum',
    'Order ID': 'count'
}).rename(columns={'Order ID': 'Order Count'})

print("Monthly trends")
print(monthly_trends.tail(), "\n\n")

#Visualization
monthly_trends.index = monthly_trends.index.to_timestamp()

# --- Sales ---
plt.figure(figsize=(8,4))
plt.plot(monthly_trends.index, monthly_trends['Sales Amount'], marker='o')
plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)
plt.show()

# --- Profit ---
plt.figure(figsize=(8,4))
plt.plot(monthly_trends.index, monthly_trends['Profit'], marker='o')
plt.title('Monthly Profit Trend')
plt.xticks(rotation=45)
plt.show()

# --- Orders ---
plt.figure(figsize=(8,4))
plt.plot(monthly_trends.index, monthly_trends['Order Count'], marker='o')
plt.title('Monthly Order Trend')
plt.xticks(rotation=45)
plt.show()
```

## Contribution Analysis

```python
# Category Contribution
category_sales = df.groupby('Category')['Sales Amount'].sum()

category_contribution = (category_sales / category_sales.sum()) * 100

print("\nCategory Contribution (%):\n")
print(category_contribution.round(2))

# Region Contribution
region_sales = df.groupby('Region')['Sales Amount'].sum()

region_contribution = (region_sales / region_sales.sum()) * 100

print("\nRegion Contribution (%):\n")
print(region_contribution.round(2))


category_contribution.plot(kind='pie', autopct='%1.1f%%', title='Category Contribution')
plt.ylabel('')
plt.show()
```

## Month - Over - Month Growth

```python
monthly_trends['Sales Growth %'] = monthly_trends['Sales Amount'].pct_change() * 100
monthly_trends['Profit Growth %'] = monthly_trends['Profit'].pct_change() * 100

print("\nMonthly Growth:\n")
print(monthly_trends[['Sales Growth %', 'Profit Growth %']].round(2))

monthly_trends['Sales Growth %'].plot(kind='line', marker='o', title='MoM Sales Growth')
plt.xticks(rotation=45)
plt.ylabel('Growth %')
plt.show()
```

```python
numerical_df = df[['Quantity','Unit Price','Discount (%)','Sales Amount','Cost Price','Profit']]

correlation_matrix = numerical_df.corr()
print("Correlation Matrix")
print(correlation_matrix,'\n\n')

plt.figure(figsize=(8,6))

sns.heatmap(correlation_matrix, 
            annot=True,      # show values
            cmap='coolwarm', # color theme
            fmt='.2f')       # 2 decimal places

plt.title('Correlation Heatmap')
plt.show()
```

## Customer Segment Analysis

```python
segment_analysis = df.groupby('Customer Segment').agg({'Sales Amount': ['sum', 'mean'], 
                                                       'Profit':['sum', 'mean'],
                                                       'Order ID': 'count',
                                                       'Discount (%)':'mean'}).round(2)
segment_analysis.columns = ['Total Sales','Avg Sales','Total Profit','Avg Profit','Order Count','Avg Discount']
print(segment_analysis,'\n\n')


plt.figure(figsize=(12,10))

# Total Sales
plt.subplot(3,2,1)
segment_analysis['Total Sales'].plot(kind='bar')
plt.title('Total Sales')

# Avg Sales
plt.subplot(3,2,2)
segment_analysis['Avg Sales'].plot(kind='bar')
plt.title('Avg Sales')

# Total Profit
plt.subplot(3,2,3)
segment_analysis['Total Profit'].plot(kind='bar')
plt.title('Total Profit')

# Avg Profit
plt.subplot(3,2,4)
segment_analysis['Avg Profit'].plot(kind='bar')
plt.title('Avg Profit')

# Order Count
plt.subplot(3,2,5)
segment_analysis['Order Count'].plot(kind='bar')
plt.title('Order Count')

# Avg Discount
plt.subplot(3,2,6)
segment_analysis['Avg Discount'].plot(kind='bar')
plt.title('Avg Discount')

plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

## Key insights and summary

```python
print("FINAL INSIGHTS:\n")

print("1. Sales are evenly distributed across categories, indicating a diversified business model.")

print("2. Electronics and Grocery are the top contributing categories to overall revenue.")

print("3. North and South regions contribute the most to sales, while West is underperforming.")

print("4. Profit margins are consistent (~25%) across categories and regions, showing stable pricing and cost structure.")

print("5. High discounts negatively impact profit, indicating a need for better discount strategies.")

print("6. Month-over-month analysis shows high volatility with significant spikes and drops, indicating seasonal or operational fluctuations.")

print("7. Sales and profit are strongly correlated, meaning revenue growth directly impacts profitability.\n")

total_sales = df['Sales Amount'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
avg_profit_margin = (total_profit/total_sales)*100

print(f'Total Sales: {total_sales:.2f}')
print(f'Total Profit: {total_profit:.2f}')
print(f'Total Orders: {total_orders:.2f}')
print(f'Average Profit Margin: {avg_profit_margin:.2f}%')

top_categories = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print(f'\n TOP PERFORMING CATEGORIES:\n')
for i, (category,profit) in enumerate (top_categories.items(),1):
    print(f'{i}.{category}: ${profit:,.2f}')


best_region = df.groupby('Region')['Profit'].sum().idxmax()
print(f'\nBest Performing Region:{best_region}')

delivery_status = df['Delivery Status'].value_counts(normalize=True)*100
print(f'\nDelivery Status\n')
for status, percentage in delivery_status.items():
    print(f'{status}:{percentage:.1f}%')

best_segment =df.groupby('Customer Segment')['Profit'].sum().idxmax()
print(f'\nMost Profitable customer segment: {best_segment}')
```
